In quantum physics a **wave function**, $\Psi$, is a mathematical description of the state of a quantum system. The wave function is a complex-valued probability amplitude, and the probabilities for the possible results of measurements made on the system can be derived from it.

For now on, let's consider the simple case of a non-relativistic single particle, without spin, in three spatial dimensions. The state of such a particle is completely described by its wave function, $\Psi(\vec{r},t)$, where $\vec{r}$ is position and $t$ is time.

For one spinless particle, if the wave function is interpreted as a probability amplitude, the square modulus of the wave function, $|\Psi(\vec{r},t)|^2$, is interpreted as the probability density that the particle is at $\vec{r}$ at time $t$. #### Hydrogen atom
The wave functions of the eigenstates of an electron in a Hydrogen atom (this is the only atom for which the Schroedinger equation has been solved exactly) are expressed in terms of spherical harmonics and generalized Laguerre polynomials. It is convenient to use spherical coordinates, and the wave function can be separated into functions of each coordinate:

$$
\Psi_{n,l,m}(r,\theta,\phi)=
\sqrt{\left(\frac{2}{na_0}\right)^3 \frac{(n-l-1)!}{2n[(n+l)!]}}
e^{-r/na_0}\left(\frac{2r}{na_0}\right)^l L_{n-l-1}^{2l+1}\left(\frac{2r}{na_0}\right)
Y_l^m(\theta,\phi)
$$

where $a_0=4\pi\epsilon_0\hbar^2/m_e e^2=0.0529$ nm is the Bohr radius, $L_{n-l-1}^{2l+1}$ are the generalized Laguerre polynomials of degree $n-l-1$, $n=1,2,...$ is the principal quantum number, $l=0,1, ..., n-1$ the azimuthal quantum number, $m=-l, -l+1, ..., l-1, l$ the magnetic quantum number.

For example, the ground state wave function is:
$$
\Psi_{1,0,0}(r,\theta,\phi)=
\frac{a_0^{-3/2}}{\sqrt{\pi}} e^{-r/a_0} 
$$
whereas one of the three $2p$ excited state is:
$$
\Psi_{2,1,0}(r,\theta,\phi)=
\frac{a_0^{-5/2}}{8}\sqrt{\frac{2}{\pi}} r e^{-r/2a_0} \cos(\theta)
$$
<span style="color:blue">Expectation values for the radius turns out to be exactly:
$$
\left\langle r \right\rangle_{\Psi_{1,0,0}} = \frac{3}{2}a_0
\quad
\left\langle r \right\rangle_{\Psi_{2,1,0}} = 5 a_0
$$
</span>

#### Metropolis

L'algoritmo di Metropolis consiste nel generare una catena di punti distribuiti secondo la densità di probabilità che si vuole campionare. Ad ogni passo dell'algoritmo viene proposto un nuovo punto che può essere accettato o rifiutato.

Per generare i punti della catena scelgo uno step $\Delta$ che permetta di ottenere un'accettazione di $\approx 50\%$: nel caso di probabilità di transizione uniforme lo estraggo distribuito come $\Delta \sim U(0,\delta)$, con $\delta=1.1$ per l'orbitale $\Psi_{100}$ e $\delta=2.8$ per l'orbitale $\Psi_{210}$; nel caso di probabilità di transizione gaussiana lo estraggo come $\Delta \sim N(0, \sigma^2)$, con $\sigma = \frac{2}{3} \delta$. In questo modo posso generare dei punti di test traslando le coordinate del punto come:

$$ x_i^{\mathrm{test}} = x_i^{\mathrm{old}} + \Delta_i .$$

Utilizzo ciascuno dei punti $\vec{x}^{\mathrm{ \ test}} = (x, y, z)^{\mathrm{test}}$ e $\vec{x}^{\mathrm{ \ old}} = (x, y, z)^{\mathrm{old}}$ per valutare le densità di probabilità:

$$ P_{1,0,0} = |\Psi_{ \ 1,0,0}|^2 \propto e^{-2r/a_0} ,$$
$$
P_{2,1,0} = |\Psi_{ \ 2,1,0}|^2 \propto
r^2 e^{-r/a_0} \cos^2(\theta) ,
$$

convertendo le coordinate cartesiane in sferiche e trascurando i fattori di normalizzazione, che non influiscono sul confronto della probabilità in punti diversi. Se le probabilità sono tali per cui $P^{\mathrm{test}}>P^{\mathrm{old}}$ il nuovo punto viene aggiunto alla catena, altrimenti lo si accetta con probabilità $P^{\mathrm{test}}/P^{\mathrm{old}}$. Se il punto viene rifiutato si aggiunge una copia di $(x, y, z)^{\mathrm{old}}$ alla catena. 

I primi punti della catena dipendono molto dalla posizione iniziale in cui si posiziona il camminatore. Pertanto scelgo una posizione abbastanza vicina al massimo atteso: $\vec{x}_0 = (1, 1, 1)$ per $\Psi_{100}$ e $\vec{x}_0 = (3, 3, 3)$ per $\Psi_{210}$, che danno rispettivamente $r_{100} \approx 1.7$ e $r_{210} \approx 5.2$.

In generale è necessario equilibrare, ossia eliminare i punti percorsi dal camminatore prima di avvicinarsi alla regione in cui la densità di probabilità ha valori significativi. Per far ciò si osserva se nel grafico della distanza $r$ in funzione dei passi della simulazione è presente un trend, che indica che non si è ancora raggiunto l'equilibrio, oppure se ci sono solo fluttazioni attorno a valori plausibili.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
r_eq_100_near = np.loadtxt("Es05.1/equilibrazione_100_near.dat", usecols=(0), delimiter=' ', unpack=True)
r_eq_210_near = np.loadtxt("Es05.1/equilibrazione_210_near.dat", usecols=(0), delimiter=' ', unpack=True)
r_eq_100_far = np.loadtxt("Es05.1/equilibrazione_100_far.dat", usecols=(0), delimiter=' ', unpack=True)
r_eq_210_far = np.loadtxt("Es05.1/equilibrazione_210_far.dat", usecols=(0), delimiter=' ', unpack=True)

fig = plt.figure(figsize = (15,12))
plt.suptitle('Equilibrazione (step uniforme)', size=20)

plt.subplot(2, 2, 1)

plt.plot(r_eq_100_near)
plt.xlabel('# steps', size=12)
plt.ylabel('$r \ [a_0]$', size=12)
plt.title(r'$r_{\Psi_{100}}$ partendo da $\vec{x}_0 = (1, 1, 1)$', size=15)
plt.grid(True)

plt.subplot(2, 2, 2)

plt.plot(r_eq_210_near)
plt.xlabel('# steps', size=12)
plt.ylabel('$r \ [a_0]$', size=12)
plt.title(r'$r_{\Psi_{210}}$ partendo da $\vec{x}_0 = (3, 3, 3)$', size=15)
plt.grid(True)

plt.subplot(2, 2, 3)

plt.plot(r_eq_100_far)
plt.xlabel('# steps', size=12)
plt.ylabel('$r \ [a_0]$', size=12)
plt.title(r'$r_{\Psi_{100}}$ partendo da $\vec{x}_0 = (10, 10, 10)$', size=15)
plt.grid(True)

plt.subplot(2, 2, 4)

plt.plot(r_eq_210_far)
plt.xlabel('# steps', size=12)
plt.ylabel('$r \ [a_0]$', size=12)
plt.title(r'$r_{\Psi_{210}}$ partendo da $\vec{x}_0 = (20, 20, 20)$', size=15)
plt.grid(True)

plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

Nella prima riga è rappresentato l'andamendo della distanza dal centro partendo da un punto vicino al massimo atteso della densità di probabilità, mentre nei grafici della seconda il punto di partenza è lontano.

Si osserva che partendo in posizione vicina al massimo si hanno solo fluttuazioni, mentre partendo da punti lontani c'è un trend del valore di $r$ che scende per avvicinarsi alla zona di interesse. Tuttavia dopo pochi passi tale valore si stabilizza, pertanto scelgo $n_{step}=200$ come numero di passi di equilibrazione, in modo da eliminare eventuali punti molto lontani dalle regioni dove la distribuzione di probabilità è massima.

Fare step gaussiani non influisce sull'andamento qualitativo di questi grafici né sul numero di step necessari per equilibrare.

#### Stima del raggio medio

Dopo aver equilibrato, utilizzo l'algoritmo di Metropolis per campionare $10^6$ punti distribuiti secondo $|\Psi_{100}|^2$ e $|\Psi_{210}|^2$, utilizzando le probabilità di transizione uniforme e gaussiana. Per valutare $\langle r \rangle$ utilizzo la tecnica della media a blocchi, suddividendo l'insieme di dati in $100$ blocchi.

```python
N, sum_prog_r_100, err_prog_r_100 = np.loadtxt("Es05.1/risultati_05.1_100.dat", usecols=(0,1, 2), delimiter=' ', unpack=True)
N, sum_prog_r_210, err_prog_r_210 = np.loadtxt("Es05.1/risultati_05.1_210.dat", usecols=(0,1, 2), delimiter=' ', unpack=True)
N, sum_prog_r_100_gauss, err_prog_r_100_gauss = np.loadtxt("Es05.1/risultati_05.1_100_gauss.dat", usecols=(0,1, 2), delimiter=' ', unpack=True)
N, sum_prog_r_210_gauss, err_prog_r_210_gauss = np.loadtxt("Es05.1/risultati_05.1_210_gauss.dat", usecols=(0,1, 2), delimiter=' ', unpack=True)



fig = plt.figure(figsize = (15,6))
plt.suptitle(' ')


plt.subplot(1, 2, 1)

plt.errorbar(N, sum_prog_r_100,yerr=err_prog_r_100, label='Step uniforme', linewidth=1, elinewidth=1, zorder=2)
plt.errorbar(N, sum_prog_r_100_gauss,yerr=err_prog_r_100_gauss, label='Step gaussiano', linewidth=1, elinewidth=1, zorder=3)
plt.hlines(1.5, 0, N[-1], zorder=1)
plt.xlabel('# steps', size=12)
plt.ylabel(r'$\langle r \rangle \ [a_0]$', size=12)
plt.title('$<r>_{\Psi_{100}}$', size=15)
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)

plt.errorbar(N, sum_prog_r_210,yerr=err_prog_r_210, label='Step uniforme', linewidth=1, elinewidth=1, zorder=2)
plt.errorbar(N, sum_prog_r_210_gauss,yerr=err_prog_r_210_gauss, label='Step gaussiano', linewidth=1, elinewidth=1, zorder=3)
plt.hlines(5, 0, N[-1],zorder=1)
plt.xlabel('# steps', size=12)
plt.ylabel(r'$\langle r \rangle \ [a_0]$', size=12)
plt.title('$<r>_{\Psi_{210}}$', size=15)
plt.grid(True)
plt.legend()


plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

Per entrambe le distribuzioni di probabilità e per entrambi i tipi di step la stima di $<r>$ converge al valore atteso compatibilmente alle incertezze, in seguito a qualche fluttuazione iniziale. Non si osservano differenze sostanziali tra il campionamento con step uniformi e quello con step gaussiani.

#### Distribuzione 3D dei punti campionati

```python
X_100, Y_100, Z_100 = np.loadtxt("Es05.1/distribuzione_100.dat", usecols=(0,1, 2), delimiter=' ', unpack=True)
X_210, Y_210, Z_210 = np.loadtxt("Es05.1/distribuzione_210.dat", usecols=(0,1, 2), delimiter=' ', unpack=True)
X_100_gauss, Y_100_gauss, Z_100_gauss = np.loadtxt("Es05.1/distribuzione_100_gauss.dat", usecols=(0,1, 2), delimiter=' ', unpack=True)
X_210_gauss, Y_210_gauss, Z_210_gauss = np.loadtxt("Es05.1/distribuzione_210_gauss.dat", usecols=(0,1, 2), delimiter=' ', unpack=True)
```

```python
fig = plt.figure(figsize = (15,12))
plt.suptitle('Distribuzioni dei punti campionati nelle simulazioni', size=20)

lim100=8
lim210=20

ax = fig.add_subplot(2, 2, 1, projection='3d')

ax.scatter(X_100, Y_100, Z_100, c=Z_100, marker='.', s=0.1)
ax.set_aspect('equal')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_xlim(-lim100,lim100)
ax.set_ylim(-lim100,lim100)
ax.set_zlim(-lim100,lim100)
ax.set_title('$|\Psi_{100}|^2$ (step uniforme)', size=15)
ax.view_init(10, 30)


ax = fig.add_subplot(2, 2, 2, projection='3d')

ax.scatter(X_210, Y_210, Z_210, c=Z_210, marker='.', s=1)
ax.set_aspect('equal')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_xlim(-lim210,lim210)
ax.set_ylim(-lim210,lim210)
ax.set_zlim(-lim210,lim210)
ax.set_title('$|\Psi_{210}|^2$ (step uniforme)', size=15)
ax.view_init(10, 30)


ax = fig.add_subplot(2, 2, 3, projection='3d')

ax.scatter(X_100_gauss, Y_100_gauss, Z_100_gauss, c=Z_100_gauss, marker='.', s=1)
ax.set_aspect('equal')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_xlim(-lim100,lim100)
ax.set_ylim(-lim100,lim100)
ax.set_zlim(-lim100,lim100)
ax.set_title('$|\Psi_{100}|^2$ (step gaussiano)', size=15)
ax.view_init(10, 30)


ax = fig.add_subplot(2, 2, 4, projection='3d')

ax.scatter(X_210_gauss, Y_210_gauss, Z_210_gauss, c=Z_210_gauss, marker='.', s=1)
ax.set_aspect('equal')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_xlim(-lim210,lim210)
ax.set_ylim(-lim210,lim210)
ax.set_zlim(-lim210,lim210)
ax.set_title('$|\Psi_{210}|^2$ (step gaussiano)', size=15)
ax.view_init(10, 30)

plt.show();
```

Le distribuzioni spaziali dei punti campionati rappresentano le densità di probabilità per gli orbitali $\Psi_{100}$ e $\Psi_{210}$. La forma dello stesso orbitale non cambia significamente se il campionamento è stato effettuato con step uniforme o gaussiano.
