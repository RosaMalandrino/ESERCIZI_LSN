```python
n_steps_s, AC_U_solid, AC_P_solid = np.loadtxt("Es07.1/Analisi/AC_solid.dat", usecols=(0,1, 2), delimiter=' ', unpack=True)
n_steps_l, AC_U_liquid, AC_P_liquid = np.loadtxt("Es07.1/Analisi/AC_liquid.dat", usecols=(0,1, 2), delimiter=' ', unpack=True)
n_steps_g, AC_U_gas, AC_P_gas = np.loadtxt("Es07.1/Analisi/AC_gas.dat", usecols=(0,1, 2), delimiter=' ', unpack=True)



fig = plt.figure(figsize = (15,5))
plt.suptitle('Fase solida', size=20)

plt.subplot(1, 2, 1)    

plt.plot(n_steps_s, AC_U_solid)
plt.xlabel("t [steps]")
plt.ylabel("AC")
plt.title("Funzione di autocorrelazione per U/N")
plt.grid(True)

plt.subplot(1, 2, 2)   

plt.plot(n_steps_s, AC_P_solid)
plt.xlabel("t [steps]")
plt.ylabel("AC")
plt.title("Funzione di autocorrelazione per P")
plt.grid(True)

plt.subplots_adjust(wspace=0.3, hspace=0.3)



fig = plt.figure(figsize = (15,5))
plt.suptitle('Fase liquida', size=20)

plt.subplot(1, 2, 1)    

plt.plot(n_steps_l, AC_U_liquid)
plt.xlabel("t [steps]")
plt.ylabel("AC")
plt.title("Funzione di autocorrelazione per U/N")
plt.grid(True)

plt.subplot(1, 2, 2)   

plt.plot(n_steps_l, AC_P_liquid)
plt.xlabel("t [steps]")
plt.ylabel("AC")
plt.title("Funzione di autocorrelazione per P")
plt.grid(True)

plt.subplots_adjust(wspace=0.3, hspace=0.3)



fig = plt.figure(figsize = (15,5))
plt.suptitle('Fase gassosa', size=20)

plt.subplot(1, 2, 1)    

plt.plot(n_steps_g, AC_U_gas)
plt.xlabel("t [steps]")
plt.ylabel("AC")
plt.title("Funzione di autocorrelazione per U/N")
plt.grid(True)

plt.subplot(1, 2, 2)   

plt.plot(n_steps_g, AC_P_gas)
plt.xlabel("t [steps]")
plt.ylabel("AC")
plt.title("Funzione di autocorrelazione per P")
plt.grid(True)

plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

By means of your upgraded MC code, equilibrate and <span style="color:red">perform MC NVT simulations via a Lennard-Jones model</span> of Argon ($\sigma = 0.34$ nm, $\epsilon/k_B = 120$ K, $m=39.948$ amu) in the following conditions:

#### Simulazioni Argon

Simulo le tre differenti fasi (solida, liquida e gas) dopo averle opportunamente equilibrate partendo da una configurazione iniziale di cristallo FCC perfetto.
Per evitare configurazioni correlate utilizzo il metodo della media a blocchi, scegliendo una taglia sufficientemente grande per evitare di sottostimare l'errore.

Eseguo anche la simulazione NVE per calcolare la funzione $g(r)$ da confrontare con quella ottenuta con il metodo Monte CArlo. Per queste simulazioni utilizzo come configurazioni di partenza quelle equilibrate per la quarta esercitazione.

Il codice utilizza le unit√† di Lennard-Jones. Per passare a unit√† del sistema internazionale √® necessario tenere conto delle seguente relazioni per l'energia, la pressione e il raggio:


$$ E = \epsilon E^* $$

$$ P = \frac{\epsilon}{\sigma^3} P^* $$

$$ r = \sigma r^* $$

```python
##### fattori di conversione #####

kB = 1.380649 * 10**-23  #[J K^-1]
T2SI = 120               #energia da unit√† di LJ a SI [K]
E2SI =  T2SI*kB          #energia da unit√† di LJ a SI [J] (=epsilon)

sigma_nm=0.34            #sigma in nm
sigma=sigma_nm*10**-9    #sigma in m

P2SI=E2SI*sigma**-3      #pressione da unit√† di LJ a SI [J m^-3]

r2SI=sigma_nm
```

### Fase solida

Simulo un sistema a $\rho^\star = 1.1$, $T^\star = 0.8$ e $r_c = 2.2$. Per entrambe le simulazioni uso $100$ blocchi, ciascuno che consiste di $1000$ step. Nel caso NVE la misura √® effettuata ogni $10$ passi.

```python
N, sum_prog_epot, err_prog_epot = np.loadtxt("Es07.4/Solid/Epot.dat", usecols=(0,2,3), delimiter=None, unpack=True)
N, sum_prog_pres, err_prog_pres = np.loadtxt("Es07.4/Solid/Pres.dat", usecols=(0,2,3), delimiter=None, unpack=True)
r, g_NVT, err_g_NVT = np.loadtxt("Es07.4/Solid/g_ave_NVT.dat", usecols=(0,1,2), delimiter=None, unpack=True)
r, g_NVE, err_g_NVE = np.loadtxt("Es07.4/Solid/g_ave_NVE.dat", usecols=(0,1,2), delimiter=None, unpack=True)


fig = plt.figure(figsize = (15,12))
plt.suptitle('Fase solida', size=20)

plt.subplot(2, 2, 1)
plt.errorbar(N,sum_prog_epot*E2SI, yerr=err_prog_epot*E2SI)
plt.xlabel('# steps', size=12)
plt.ylabel('$E_{pot}$ [$ \ J$]', size=15)
plt.title('Energia potenziale', size=18)
plt.grid(True)


plt.subplot(2, 2, 2)
plt.errorbar(N,sum_prog_pres*P2SI, yerr=err_prog_pres*P2SI)
plt.xlabel('# steps', size=15)
plt.ylabel('$P$ [$\ J m^{-3}$]', size=15)
plt.title('Pressione', size=18)
plt.grid(True)


plt.subplot(2, 1, 2)

plt.errorbar(r*r2SI, g_NVT, yerr=err_g_NVT, label='MonteCarlo NVT')
plt.errorbar(r*r2SI, g_NVE, yerr=err_g_NVE, label='MolDyn NVE')
plt.xlabel('$r$ [nm]', size=15)
plt.ylabel('$g(r)$', size=15)
plt.title('Funzione di distribuzione radiale (NVT)', size=18)
plt.grid(True)
plt.legend()


plt.subplots_adjust(wspace=0.3, hspace=0.3)
```
