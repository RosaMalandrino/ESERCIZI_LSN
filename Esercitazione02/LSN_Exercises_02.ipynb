#### Integrazione Monte Carlo

Supponiamo di voler calcolare l'integrale:
$$ I = \int_a^b f(x) dx.$$

Possiamo pensare di suddivide l'intervallo $\left[ a, b \right]$ in $N$ sottointervalli di dimensione $\Delta x$ e approssimare l'area sottesa dalla curva con la somma dei rettangoli determinati da questa partizione. Al crescere del valore di $N$ questa stima si avvicinerà sempre di più al valore vero dell'integrale fino a coincidervi:
$$ I = \int_a^b f(x) dx = \lim_{N \to \infty} \sum_{i=1}^N \Delta x \cdot f(x_i),$$

dove i valori $x_i$ sono i punti medi di ciascun intervallo. Scrivendo l'intervallo $\Delta x$ come suddivisione dell'intervallo $\left[a, b \right]$ in $N$ parti uguali si ottiene:
$$ I=\lim _{N \rightarrow \infty} \sum_{i=1}^{N} \Delta x \cdot f\left(x_{i}\right)=(b-a) \lim _{N \rightarrow \infty} \frac{1}{N} \sum_{i=1}^{N} f\left(x_{i}\right)=(b-a)\langle f\rangle_{[a, b]}, $$

dove con $\langle f\rangle_{[a, b]}$ si intende la media dei valori $f(x_i)$.
Campionando dunque $N$ punti distribuiti uniformemente nell'intervallo $\left[a, b \right]$, è possibile ottenere una stima dell'integrale come:
$$ I \simeq (b-a) \frac{1}{N} \sum_{i=1}^{N} f\left(x_{i}\right). $$

#### Importance sampling

In generale si vuole valutare un integrale della forma:
$$ I = \int_{\Omega} g(x) p(x) dx , $$

dove $p(x)$ è una distribuzione di probabilità definita positiva sul dominio $\Omega$ e normalizzata. In questo caso possiamo estrarre un insieme di variabili $\left\{ x_1, ..., x_N \right\}$ distribuite con densità di probabilità $p(x)$ e stimare l'integrale come:
$$ I \simeq G_{N}=\frac{1}{N} \sum_{i=1}^{N} g\left(x_{i}\right). $$


Il metodo dell'importance sampling o campionamento rilevante consiste nel fattorizzare l'integranda come il prodotto tra una funzione e un'opportuna distribuzione di probabilità:
$$ I = \int_{\Omega} g(x) p(x) dx = \int_{\Omega} \frac{g(x) p(x)}{q(x)} q(x) dx = \int_{\Omega} f(x) q(x) dx .$$

La distribuzione $q(x)$ deve essere definita positiva sul dominio di integrazione e normalizzata; inoltre la funzione $\frac{g(x) p(x)}{q(x)}$ deve essere limitata ovunque a meno di un insieme di punti a misura nulla. Questo procedimento è utile perché la distribuzione $p(x)$ può spesso essere sostituita da una scelta migliore $q(x)$ che rende il calcolo più preciso.
Infatti si può dimostrare che quanto più la distribuzione $q(x)$ è simile all'integranda originale $g(x)p(x)$, tanto migliore è la stima dell'integrale e dunque tanto più piccola è la varianza $\sigma^2$.


####   Calcolo dell'integrale 

L'integrale da calcolare e il seguente:

$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1.$$

$$ I = \int_0^1 f(x) dx = \int_0^1 \frac{f(x)}{r(x)} r(x) dx ,$$

dove $r(x)$ è una distribuzione normalizzata scelta opportunamente in modo che approssimimi l'integranda. In particolare scelgo la retta che taglia il coseno: $r(x) = 2 (1- x)$, definiendola solo nell'intervallo $\left[ 0,1 \right]$.

Per stimare l'integrale nei due modi genero per ciascuna delle due densità di probabilità $M=10^6$ variabili casuali. Per avere numeri distribuiti secondo $r(x)$ utilizzo il metodo della comulativa, calcolando:
$$y=F_x(x)= \int_0^x r(x)dx=2x- x^2 \quad \quad \quad \quad \quad \mathrm{con} \ x<1$$ e invertendo. Se $y$ è una variabile distribuita uniformemente tra $0$ e $1$, allora $x=1-\sqrt{1-y} \ $ è una variabile con densità di probabilità $r(x)$ nello stesso intervallo.

Utilizzo il metodo della media a blocchi con $N=100$ per valutare l'errore progressivo dei due integrali.

Si osserva come entrambi i metodi convergano al valore esatto $I=1$ con un'incertezza che decresce progressivamente all'aumentare dei blocchi $N$. Tuttavia rappresentando i due andamenti sulla stessa scala è evidente come l'integrale calcolato con l'importance sampling con distribuzione $r(x)$ converga al risultato con un errore inferiore rispetto a quello calcolato con il campionamento uniforme, in particolare nell'ultimo blocco tale incertezza è ridotta di un fattore $\approx 4$.

Infine nel grafico in basso a destra sono rappresentate l'integranda $f(x)$ e le due densità di probabilità utilizzate per il campionamento di punti. È quindi evidente come la distribuzione che approssima meglio la funzione $f(x)$ porti a una migliore precisione nella stima del suo integrale.

#### Random walk e diffusione

Il random walk è un tipo di traiettoria che consiste in una serie di passi consecutivi casuali, ciascuno dei quali avviene seguendo una certa distribuzione di probabilità. Esso è un esempio di processo di Markov, dal momento che la probabilità di trovarsi in un dato stato $(x_n, t_n)$ dipende solo dallo stato immediatamente precedente $(x_{n-1}, t_{n-1})$:
$$ p_{1 | n-1}\left(x_{n}, t_{n} | x_{1}, t_{1} ; \ ... \ ; x_{n-1}, t_{n-1}\right)=p_{1|1}\left(x_{n}, t_{n} | x_{n-1}, t_{n-1}\right) .$$

Non viene quindi conservata memoria di tutti gli $n-1$ stati passati. La forma della probabilità $p_{1|1}\left(x_{n}, t_{n} | x_{n-1}, t_{n-1}\right)$ dipende delle caratteristiche del processo considerato.



I passi avvengono a tempi discreti con intervallo $\tau$ e hanno lunghezza finita che indichiamo con $a$. Per studiare il comportamento del random walk si può pensare di considerare il caso monodimensionale e porre $a \to 0$ e $\tau \to 0$. In questo modo si ottiene una distribuzione di probabilità di trovare il camminatore in una posizione $x$ del continuo ad un qualsiasi tempo $t$. Si può dimostrare che l'equazione differenziale che ne risulta è l'equazione di diffusione con coefficiente $D=\frac{a^2}{2\tau}$. La sua soluzione dà la distribuzione di probabilità:

$$ p_{1}(x, t)=\frac{1}{\sqrt{4 \pi D t}} \exp{\left[-\frac{x^{2}}{4 D t} \right]} $$

una gaussiana con media nulla e $\sigma^2 = 2Dt $.

Si può osservare che la distribuzione somma $S_N = \sum x_i$ data dal teorema del limite centrale consiste nella posizione finale di un random walk composto da $N$ passi, in cui ogni variabile $x_i$ corrisponde a un passo con una determinata densità di probabilità. Quest'ultima è rappresentata da $p(x)=0.5[\delta(x+a)+\delta(x-a)]$, che ha media nulla e varianza $\sigma^2 = a^2$. Di conseguenza $S_N$ per $N \to \infty$ tenderà a una distribuzione gaussiana con media pari a $0$ e varianza $\sigma^2_{S_N} = N \sigma^2 = N a^2$. Confrontando con la varianza della distribuzione ottenuta dal random walk si può osservare che coincidono.

Di conseguenza il comportamento diffusivo del random walk può essere rivelato dalla forma della deviazione standard, che all'aumentare di $N$ segue la legge: $\sigma = a \sqrt{N}$.

La trattazione fatta in una dimensione può essere estesa al caso $d$-dimensionale considerando la distanza dal centro $|\vec{r}|$ al posto di $x$.

####  Simulazione random walk

In particolare considero due casi:
1. random walk discreto, in cui ad ogni passo il camminatore si sposta in un punto adiacente su un reticolo tridimensionale caratterizzato da una spaziatura $a$;
2. random walk continuo, in cui la lunghezza del passo è fissata ad $a$, ma esso può avvenire in qualsiasi direzione dello spazio $\mathbb{R}^3$.


Nel caso discreto campiono il passo generando casualmente una variabile con sei possibili risultati, a ciascuno dei quali viene assegnata una direzione di movimento e il verso.
Nel caso continuo invece per determinare un punto sulla superficie della sfera di raggio $a$ estraggo casualmente dei punti al suo interno, da cui ricavo le rispettive coordinate sferiche $\theta$ e $\phi$. In entrambi i casi pongo $a=1$.

Considero un random walk con numero di passi pari a $n_{steps} = 100$ e lo ripeto $M=10^5$ volte sia per il caso sul reticolo cubico che per quello nello spazio libero. Divido queste simulazioni in $100$ blocchi per poter calcolare l'errore con il metodo della media a blocchi.

Per entrambi i random walk stimo la grandezza $\langle |\vec{r}_N|^2 \rangle$ in funzione dell'N-esimo passo, con $N=1, ..., n_{steps}$, da cui estraggo la radice $\sqrt{\langle |\vec{r}_N|^2 \rangle}$. Questa grandezza rappresenta il valore quadratico medio della posizione del camminatore, calcolato rispetto al valor medio della traiettoria $\langle r \rangle = 0$. Considerando la distribuzione di probabilità del random walk ci si aspetta il seguente comportamento: $\sqrt{\langle |\vec{r}_N|^2 \rangle} = \sigma \propto \sqrt{N}$, pertanto posso fittare le stime ottenute con una funzione della forma: $f(N)=k\sqrt{N}$.

Per quanto riguarda invece l'errore su $\sqrt{\langle |\vec{r}_N|^2 \rangle}$ esso può essere valutato propagando l'errore sulla stima di $ \langle |\vec{r}_N|^2 \rangle$ ottenuto con il metodo delle medie a blocchi attraverso la funzione di radice quadrata.

```python
N, r_d, err_d , r_c, err_c = np.loadtxt("Es02.2/risultati_02.2.dat", usecols=(0,1,2,3,4), delimiter=' ', unpack=True)

#fit

def f(N, k):
    
    return k*np.sqrt(N);

p_opt_d, p_cov_d = curve_fit(f, N, r_d)
y_fit_d = f(N,p_opt_d[0])
p_opt_c, p_cov_c = curve_fit(f, N, r_c)
y_fit_c= f(N,p_opt_c[0])


fig = plt.figure(figsize = (15,15))
plt.suptitle('Random walk', size=20, y=0.95)

plt.subplot(2, 2, 1)
plt.errorbar(N,r_d,yerr=15*err_d, errorevery=2, capsize=2, marker='o', ms=2, markevery=2, ls='', elinewidth=1, label='Dati con barre d\'errore a $15\sigma$')
plt.plot(N,y_fit_d, label='Fit $f(N) = k\sqrt{N}$')
plt.xlabel('# passi', size=12)
plt.ylabel(r'$\sqrt{\langle |\vec{r}^2| \rangle}$', size=12)
plt.title('RW su reticolo cubico', size=15)
plt.grid(True)
plt.legend(loc=4)

plt.subplot(2, 2, 2)
plt.errorbar(N,r_c,yerr=15*err_c, errorevery=2, capsize=2, marker='o', ms=2, markevery=2, ls='', elinewidth=1, label='Dati con barre d\'errore a $15\sigma$')
plt.plot(N,y_fit_c, label='Fit $f(N) = k\sqrt{N}$')
plt.xlabel('# passi', size=12)
plt.ylabel(r'$\sqrt{\langle |\vec{r}^2| \rangle}$', size=12)
plt.title('RW sullo spazio continuo', size=15)
plt.grid(True)
plt.legend(loc=4)

plt.subplot(2, 2, 3)
plt.hlines(0, 0, N[-1], color='orange', linestyle='--', lw=2, label='Fit')
plt.errorbar(N,r_d-y_fit_d,yerr=err_d, capsize=2, marker='o', ms=3, mew=0.5, linewidth=0.5, elinewidth=1, label='Dati con barre d\'errore a $1\sigma$')
plt.xlabel('# passi', size=12)
plt.ylabel(r'$\Delta r$', size=12)
plt.title('Differenze (reticolo cubico)', size=15)
#plt.ylim(-0.025, 0.025)
plt.grid(True)
plt.legend(loc=2)

plt.subplot(2, 2, 4)
plt.hlines(0, 0, N[-1], color='orange', linestyle='--', lw=2, label='Fit')
plt.errorbar(N,r_c-y_fit_c,yerr=err_c, capsize=2, marker='o', ms=3, mew=0.5, linewidth=0.5, elinewidth=1, label='Dati con barre d\'errore a $1\sigma$')
plt.xlabel('# passi', size=12)
plt.ylabel(r'$\Delta r$', size=12)
plt.title('Differenze (spazio continuo)', size=15)
#plt.ylim(-0.025, 0.025)
plt.grid(True)
plt.legend(loc=2)

plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

Nei grafici in alto sono riportati gli andamenti delle stime della dispersione con le relative incertezze.
Viene rappresentata una barra d'errore ogni due punti, ciascuna con un fattore moltiplicativo arbitrario in modo da renderla visibile.
Sono riportati anche i fit ottenuti seguendo una relazione: $r_{fit}=f(N)=k \sqrt{N}$.

Per verificare la bontà del fit sottraggo ai dati i valori della curva ottenuta, rappresentando la grandezza $\Delta r$ che definisco come: $\Delta r = \sqrt{\langle |\vec{r}_N|^2 \rangle} - r_{fit}$.
Si osserva che i dati sono compatibili con il valore del fit entro $1 \sigma$, pertanto si può affermare che i risultati mostrano un comportamento diffusivo.
